```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Para facilitar a criação de distribuição de frequência nas questões, implementei essa função que retorna um dataframe com as frequências, semelhante a função disponibilizada pelo professor no notebook da semana 7.

```python
def dfrequencia(var):
    #Calcula a frequência absoluta
    var = var.values.flatten()
    absoluta = pd.value_counts(var)
    #Calcula a frequência relativa
    relativa = np.around((absoluta/len(var))*100,decimals=2)#Duas casas decimais
    #Calcula a frequência acumulada, com uma função numpy
    acumulada = np.cumsum(absoluta).astype(int)
    #Retorna a tabela pronta
    tabela = pd.DataFrame({ 'Fi': absoluta,
               'Fr': relativa,
               'Acumulada': acumulada
            })
    return tabela
```

**(c) crie uma tabela de contingência relacionando as variáveis Bairro e Tipo e outra relacionando Bairro e Nº
de quartos e comente que informações podem ser extraídas dessas tabelas.**

```python
cont_bairro_tipo = pd.crosstab(df2['Bairro'], 
                            df2['Tipo'],  
                               margins = False) 
cont_bairro_tipo
```

Com essa tabela podemos ver o tipo de residências a venda que prevalecem em cada bairro. Podemos perceber que o número de apartamentos a venda é maior do que as outras residências na maioria dos bairros, com exceção do Centro, São Conrado e Recreio.

```python
cont_bairro_quartos = pd.crosstab(df2['Bairro'], 
                            df2['N de quartos'],  
                               margins = False) 
cont_bairro_quartos
```

Nessa tabela podemos ver o número de quartos em cada residência à venda nos bairros fornecidos. Analisando, vemos que a Barra possui mais residências com variação de escolha no número de quartos, e que nos outros bairros a maioria das casas à venda possuem 3 ou mais que 4 quartos.
