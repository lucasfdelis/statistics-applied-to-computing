# 3ª Tarefa

**Para cada indicação abaixo, construa um novo dataframe, salvando-o com a terminação .csv.**

**(a) ranking de acidentes por estado para ambos os anos (um só arquivo combinando a
informação dos dois anos);**

Para fazer essa, eu usei o comando "value_counts()", e atribui o valor das planilhas a duas variáveis, que foram "acidente2019" e acidente2020". Depois, organizei elas usando a variável "dados".

A seguir, transformei os dados em um Dataframe. O comando ficou comprido mas vou explicar cada coisa que ele faz:

<h6>pd.DataFrame(dados):</h6> É o comando padrão, que transforma os dados em um Dataframe. Porém, se eu usasse ele sozinho, o index dele ficaria errado, pois ao invés de números, ele consideraria as ufs como fonte de classificação:
<img src='https://raw.githubusercontent.com/lucasfdelis/statistics-applied-to-computing/main/img4.png'>
O index dele não fica composto por números, então na hora do carregamento para uma planilha .csv, aconteceria esse erro:
<img src='https://raw.githubusercontent.com/lucasfdelis/statistics-applied-to-computing/main/img5.png'>
O .csv não reconheceria o formato do dataframe de forma correta, então apagaria o index, ou seja, a coluna que identifica os ufs.
<h6>.rename_axis('UF').reset_index()</h6>
Esse comando transforma o index em números. Assim, o .csv fica no formato correto, mantendo a coluna de ufs. O comando "rename_axis" serve para renomear o index anterior para "UF", se eu não usasse ele, o nome da coluna seria "index", que é o padrão, e isso causaria uma desorganização na planilha.
<h6>.sort_values(by = ['2019','2020']).loc[::-1,:]</h6>
Esses comandos servem para organizar a planilha e o dataframe em ordem decrescente, exibindo assim, o ranking de números de acidentes por uf.

```python
acidente2019 = df2019['uf'].value_counts()
acidente2020 = df2020['uf'].value_counts()
dados = {'2019':acidente2019,
         '2020':acidente2020}
acidentes = pd.DataFrame(dados).rename_axis('UF').reset_index().sort_values(by = ['2019','2020']).loc[::-1,:]
acidentes.to_csv('ranking2019-2020.csv',sep=';',index = None, header=True)
acidentes
```

**(b) qual a proporção de pessoas ilesas e de feridos graves por mês em cada um dos anos
considerados? (proporção = numero de ilesos ou feridos graves/ número total de
pessoas envolvidas no acidente);**

Para fazer a questão, transformei a coluna de datas para o formato de datas (comando da primeira linha), já que ela era uma string. Assim, posso manipulá-la de forma eficaz.

Então, agrupei os dados por mês e retornei a soma dos ilesos, feridos graves e número de pessoas envolvidas nos acidentes com o comando "groupby".

```python
df2019['data_inversa'] = pd.to_datetime(df2019['data_inversa'])
ilesos2019 = df2019.groupby(df2019['data_inversa'].dt.month)['ilesos'].sum()
graves2019 = df2019.groupby(df2019['data_inversa'].dt.month)['feridos_graves'].sum()
pessoas2019 = df2019.groupby(df2019['data_inversa'].dt.month)['pessoas'].sum()
```

Depois, é só fazer o cálculo de proporção, que apresentará uma lista correspondente a cada mês.

Repare que a numeração das linhas corresponde aos meses, então 1=Janeiro, 2=Fevereiro, etc.

Proporção de pessoas ilesas e de feridos graves por mês no ano de 2019:

```python
(ilesos2019+graves2019)/pessoas2019
```

Após isso, é só repetir os comandos para adquirir os dados do ano de 2020:

```python
df2020['data_inversa'] = pd.to_datetime(df2020['data_inversa'])
ilesos2020 = df2020.groupby(df2020['data_inversa'].dt.month)['ilesos'].sum()
graves2020 = df2020.groupby(df2020['data_inversa'].dt.month)['feridos_graves'].sum()
pessoas2020 = df2020.groupby(df2020['data_inversa'].dt.month)['pessoas'].sum()
```

Proporção de pessoas ilesas e de feridos graves por mês no ano de 2020:

```python
(ilesos2020+graves2020)/pessoas2020
```

**(c) mostre, reorganizando o conjunto de dados, se a pandemia conseguiu diminuir ou não
a incidência de acidentes nas rodovias federais.**

Nessa atividade, usei a mesma estrutura de código utilizadas em questões anteriores, porém, retirei a parte de ordenacão quando transformei os dados em um dataframe. Eu também contei o número de acidentes por mês com o comando "dt.month.value_counts()"

```python
df2019ord = df2019.sort_values(by='data_inversa')
df2020ord = df2020.sort_values(by='data_inversa')
ano2019 = df2019ord['data_inversa'].dt.month.value_counts()
ano2020 = df2020ord['data_inversa'].dt.month.value_counts()
dados = {'2019':ano2019,
         '2020':ano2020}
totalacidentes = pd.DataFrame(dados)
totalacidentes
```

Resolvi mostrar os dados em forma de gráfico para a visualização ficar melhor:

```python
%matplotlib inline
totalacidentes.plot.bar()
```

Com o gráfico, podemos ter certeza que a pandemia influenciou no número de acidentes registrados. Nos meses de janeiro e fevereiro, onde não havia pandemia, não houve queda. Pelo contrário, o número de acidentes foi maior que os registrados em 2019. Mas nos outros meses, com exceção de outubro,o número de acidentes em 2020 foi menor que os registrados em 2019.

Houve queda acentuada em março e abril, que foi onde as normas de circulação foram mais rígidas aqui no país, mas podemos notar também que, apesar do número de acidentes de 2020 continuarem baixos, houve um aumento de acidentes a partir do mês de maio, que corresponde a época que o lockdown acabou em várias regiões do Brasil.
